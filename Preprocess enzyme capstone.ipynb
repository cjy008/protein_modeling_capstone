```python
map_enzyme_id_to_enyzme_class = {}
id_to_sequence ={}
id_to_class = {}

id = 0
non_enzyme_data = open("non_enzyme_new_data_sequence.txt", "r")
for x in non_enzyme_data:
    sequence = x.split('>')[1].strip()
    map_enzyme_id_to_enyzme_class[id] = 0
    id_to_sequence[id] = sequence
    id +=1
```

```python
enzyme_data = open("new_data_label_sequence.txt", "r")
for x in enzyme_data:
    sequence = x.split('>')[1].strip()
    label =  x.split('>')[0][1:]
    id_to_class[id] = label
    temp_id = str(id) + "_" + label
    map_enzyme_id_to_enyzme_class[temp_id] = 1
    id_to_sequence[id] = sequence
    id_to_class[id] = label
    id +=1
```

```python
len(map_enzyme_id_to_enyzme_class)
```

```python
f = open("output_enzyme_nonenzyme_classification.fasta", "w")

for id in id_to_sequence:
    temp_id = ''
    if id in map_enzyme_id_to_enyzme_class:
        temp_id = id
    else:
        temp_id = str(id) + "_" + str(id_to_class[id])
    f.write(">" + str(temp_id) + "\n")
    f.write(id_to_sequence[id] + "\n")
f.close()
```
