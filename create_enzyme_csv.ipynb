```python
import pandas as pd
```

```python
labels = []
sequences = []
```

```python
f = open("new_data_label_sequence.txt", "r")
for x in f:
    
    splits = x.split('>')
    sequence = splits[1].strip()
    label = splits[0][1:]
    
    labels.append(label)
    sequences.append(sequence)
    
    
```

```python
sequences
```

```python
labels
```

```python
enzyme_data = pd.DataFrame({'label' : labels, 'sequence': sequences})
```

```python
enzyme_data.head()
```

```python
enyzme_class = []
enzyme_subclass = []


for label in labels:
    print(label)
    splits = label.split('.')
    enyzme_class.append(splits[0])
    enzyme_subclass.append(splits[1])
    
```

```python
enyzme_class
```

```python
enzyme_subclass
```

```python
enzyme_data['enzyme_class'] = enyzme_class
```

```python
enzyme_data['enzyme_subclass'] = enzyme_subclass
```
