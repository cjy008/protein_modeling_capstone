```python
import json
import pandas as pd
```

```python
with open('DEEPre_Sequences.json') as f:
  data = json.load(f)
```

```python
data
```

```python
df = pd.read_csv('ECPredEnzyme.csv','\t')
```

```python
df['class'] = df['EC Number'].str.split('.',expand=True)[0]
df['subclasses'] = df['EC Number'].str.split('.', expand=True)[1]
df['subclass'] = df[['class', 'subclasses']].agg('.'.join, axis=1)
df = df.drop(['subclasses'], axis=1)
```

```python
df
```

```python
class_labels = {}
subclass_labels = {}
enzymes = {}

for index, row in df.iterrows():
#     print(row)
    enzymes[row['Protein UniProt Acc.']] = "1"
    class_labels[row['Protein UniProt Acc.']] = str(row['class'])
    subclass_labels[row['Protein UniProt Acc.']] = str(row['subclass'].split('.')[1])
#     print(row['Protein UniProt Acc.'])
#     print(row['class'])
#     print(row['subclass'].split('.')[1])
#     print(row['Protein UniProt Acc.'], row['class'], row['subclass'])
```

```python
class_labels
```

```python
with open("DEEPree_class_labels.json", "w") as outfile: 
    json.dump(class_labels, outfile, indent=4)
```

```python
with open("DEEPree_subclass_labels.json", "w") as outfile: 
    json.dump(subclass_labels, outfile, indent=4)
```

```python
with open("DEEPree_enzymes_labels.json", "w") as outfile: 
    json.dump(enzymes, outfile, indent=4)
```
